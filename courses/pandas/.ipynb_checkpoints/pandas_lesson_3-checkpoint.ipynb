```python
import pandas as pd

flights = pd.read_csv("flights.csv")
flights.head()
```

```python
# .describe() function in pandas is similar to summary() in R
# can call .describe() on the whole DataFrame or
# can call .describe() on a single column within DataFrame (i.e., flights.year.describe() )

flights.describe()
flights.tailnum.describe()
```

```python
flights.head()
```

```python
# calling average on a column
# if calling .mean() on entire DataFrame, it crashes computer

flights.dep_time.mean()
```

```python
flights.dep_time.mean()
```

```python
# for non-numeric data, you *might* be interested in unique names
# similar unique() in tidyverse

flights.carrier.unique()
```

```python
# the equivalent of group_by() %>% tally(sort = TRUE) in Tidyverse

flights.carrier.value_counts()
flights.dest.value_counts()
```

```python
#flights.head()

# understand .map() function, but not sure what `lambda p: p` does

flights_distance_mean = flights.distance.mean()
flights_distance_mean

flights.distance.map(lambda p: p - flights_distance_mean)

```

```python
# create a variable showing highest air_time-to-distance ratio

ratio = (flights.air_time / flights.distance).idxmax()

which_origin = flights.loc[ratio, 'origin']

which_origin
```

```python
flights.origin.value_counts()

# Counting the number of times EWR appears in the 'origin' column
n_ewr = flights.origin.map(lambda ori: "EWR" in ori).sum()
n_ewr
n_lga = flights.origin.map(lambda ori: 'LGA' in ori).sum()
n_lga

descriptor_counts = pd.Series([n_ewr, n_lga], index=['EWR', 'LGA'])
descriptor_counts
```
