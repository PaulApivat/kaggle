```python
import pandas as pd

flights = pd.read_csv("flights.csv")
flights.head()
```

```python
# Replicate what value_counts does with groupby()

# value_counts() sorts automatically
# flights.carrier.value_counts()

flights.groupby('carrier').carrier.count()
```

```python
# after groupby() you can use any 'summary function' i.e., min()
# for operation on another column

# find the shortest distance for *each* carrier
flights.groupby('carrier').distance.min()
```

```python
# use the apply() method on a groupby()

# groupby carrier, then for each grouping, find the tailnum (with index 1)
# the index can be changed to any number
flights.groupby('carrier').apply(lambda df: df.tailnum.iloc[1])
```

```python
# groupby two columns, then .apply() to find largest hour per combination

flights.groupby(['carrier', 'tailnum']).apply(lambda df: df.loc[df.hour.idxmax()])
```

```python
# us agg() to run statistical summaries after groupby()
# equivalent to group_by() %>% summarize() in tidyverse

flights.groupby(['origin']).distance.agg([len, min, max])
```

```python
from siuba import _, group_by, summarize
from siuba.data import mtcars

(flights
    >> group_by(_.origin)
    >> summarize(avg_distance = _.distance.mean())
)
```

```python
(flights
     >> group_by(_.origin)
     >> summarize(len_distance = _.distance.count(),
                  max_distance = _.distance.max(), 
                  min_distance = _.distance.min())
)
```

```python
flights.head()

# multi-index (multiple levels) when groupby() two columns to find count (len) of another column
multi_index_example = flights.groupby(['carrier', 'origin']).distance.agg([len])
multi_index_example
```

```python
(flights
     >> group_by(_.carrier, _.origin)
     >> summarize(len = _.distance.count())
)
```

```python
# when you reset multi_index for converting back to regular index
# its what you get in default with siuba/dplyr

multi_index_example.reset_index()
```

```python
# sorting

back_to_single_index = multi_index_example.reset_index()
back_to_single_index.sort_values(by='len', ascending=False)
back_to_single_index.sort_index()
```

```python
# sorting (arrange) in siuba
from siuba import _, group_by, summarize, arrange
from siuba.data import mtcars


# arrange descending in siuba is adding negative sign i.e., arrange(-_.len)
# can use .sort_index() with siuba wrapped code!!!

(flights
     >> group_by(_.carrier, _.origin)
     >> summarize(len = _.distance.count())
     >> arrange(_.len)
).sort_index()
```

```python
# Sort by more than one column

back_to_single_index.sort_values(by=['carrier', 'len'])
```

```python
# sort (arrange) by more than one column in siuba
(flights
     >> group_by(_.carrier, _.origin)
     >> summarize(len = _.distance.count())
     >> arrange(_.carrier, _.len)
)
```
