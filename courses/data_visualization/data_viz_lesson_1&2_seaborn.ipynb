```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

players_20 = pd.read_csv("players_20.csv")
players_20.head()
players_20.tail()
```

```python
# Using Seaborn lineplot to plot age

sns.lineplot(data = players_20['age'])
```

```python
# plotting player's height in cm

sns.lineplot(data = players_20['height_cm'])
```

```python
# plotting player's weight in kg

plt.figure(figsize=(14,6))

sns.lineplot(data = players_20['weight_kg'])
```

```python
players_20.head()

# Print Name of Columns (does not show all columns)
players_20.columns

# List out ALL column names
list(players_20.columns)
```

```python
# When you plot two columns (both numeric for lineplot()), 
# seaborn automatically gives them different colors
# plots automatically counts the rows?

# determine plot size
plt.figure(figsize=(14,6))

# title of plot
plt.title("Plotting Player's Age and Height")

# plot first line - player's age
sns.lineplot(data = players_20['age'], label = "Player's Age")

# plot second line - player's height
sns.lineplot(data = players_20['height_cm'], label = "Player's Height")

# x-axis label
plt.xlabel('Number of Rows')
```
