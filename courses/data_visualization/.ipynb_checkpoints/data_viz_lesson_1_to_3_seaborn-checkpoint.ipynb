```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

players_20 = pd.read_csv("players_20.csv")
players_20.head()
players_20.tail()
```

```python
# Using Seaborn lineplot to plot age

sns.lineplot(data = players_20['age'])
```

```python
# plotting player's height in cm

sns.lineplot(data = players_20['height_cm'])
```

```python
# plotting player's weight in kg

plt.figure(figsize=(14,6))

sns.lineplot(data = players_20['weight_kg'])
```

```python
players_20.head()

# Print Name of Columns (does not show all columns)
players_20.columns

# List out ALL column names
list(players_20.columns)
```

```python
# When you plot two columns (both numeric for lineplot()), 
# seaborn automatically gives them different colors
# plots automatically counts the rows?

# determine plot size
plt.figure(figsize=(14,6))

# title of plot
plt.title("Plotting Player's Age and Height")

# plot first line - player's age
sns.lineplot(data = players_20['age'], label = "Player's Age")

# plot second line - player's height
sns.lineplot(data = players_20['height_cm'], label = "Player's Height")

# x-axis label
plt.xlabel('Number of Rows')
```

```python
import pandas as pd
import numpy as np
import datetime

# Create new column with year of date field "dob"

subset_players_20['year'] = pd.DatetimeIndex(subset_players_20['dob']).year
subset_players_20.head()
```

```python
subset_players_20_year = subset_players_20[['year', 'age', 'height_cm', 'weight_kg']]
subset_players_20_year.head()
```

```python
# set plot dimensions
plt.figure(figsize=(12,6))

sns.lineplot(data = subset_players_20_year, x='year', y='height_cm')
sns.lineplot(data = subset_players_20_year, x='year', y='weight_kg')
sns.lineplot(data = subset_players_20_year, x='year', y='age')

# add title
plt.title("Plotting players Age, Height, Weight with Year as X-axis")
```

```python
# Barchart & Heatmaps

heat_flight = pd.read_csv('heat_flight.csv')
heat_flight.head()
```

```python
# Barchart for average delay for UA carrier

plt.figure(figsize=(10,6))

plt.title("Average Delay Times for United Airlines")

sns.barplot(x = heat_flight['month'], y = heat_flight['UA'])
# this also works --> sns.barplot(x = heat_flight.month, y = heat_flight.UA)

plt.xlabel("Month")
```

```python
# Heatmap

plt.figure(figsize=(14,7))

plt.title("Average Delay by Carrier")

# annot = ensures value in each cell appears on the chart
sns.heatmap(data=heat_flight, annot=True)

plt.xlabel("Airline Carrier")
```
