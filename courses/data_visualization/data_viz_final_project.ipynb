```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

candy = pd.read_csv("candy-data.csv")
candy.head()
```

```python
ncaa = pd.read_csv("538ratingsMen.csv")
ncaa.head()
#ncaa.tail()
```

```python
# exploring line plot with x and y axis specified

sns.lineplot(x = ncaa['Season'], y = ncaa['538rating'])

```

```python
# Barplot
plt.figure(figsize=(14,8))

chart = sns.barplot(x = ncaa['TeamName'], y = ncaa['538rating'])

# x-axis orientation
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
```

```python
# filter ncaa dataframe by Season == 2019
# then re-do barplot

is_2019 = ncaa['Season']==2019
#print(is_2019.head())

# filter using boolean value
new_ncaa = ncaa[is_2019]
```

```python
# Re-plot Bar Plot using filter (newly created variable new_ncaa)
plt.figure(figsize=(16,8))

chart = sns.barplot(x = new_ncaa['TeamName'], y = new_ncaa['538rating'])

# x-axis orientation
chart.set_xticklabels(chart.get_xticklabels(), rotation=45)
```
