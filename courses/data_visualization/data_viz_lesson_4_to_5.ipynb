```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

players_20 = pd.read_csv('players_20.csv')
players_20.head()
```

```python
# Scatter Plot

plt.figure(figsize=(5,5))
sns.scatterplot(x=players_20['height_cm'], y=players_20['weight_kg'])
```

```python
# Scatter Plot with Regression line
# in ggplot = geom_smooth(method = 'lm')

plt.figure(figsize=(8,8))

sns.regplot(x=players_20['height_cm'], y=players_20['weight_kg'])
```

```python
# Add color (by Club)

plt.figure(figsize=(8,8))

sns.scatterplot(x=players_20['height_cm'], y=players_20['weight_kg'], hue=players_20['club'])
```

```python
## Distributions

# height distribution - histogram
sns.distplot(a = players_20['height_cm'], kde=False)
```

```python

# weight distribution - histogram

sns.distplot(a = players_20['weight_kg'], kde=False)
```

```python
# Density plot (same as setting kde=True in histogram)

sns.kdeplot(data = players_20['age'], shade=True)
```

```python
# 2D KDE plots - height and weight

sns.jointplot(x = players_20['height_cm'], y = players_20['weight_kg'], kind='kde')
```

```python
# 2D KDE plot - weight & age

sns.jointplot(x = players_20['age'], y = players_20['weight_kg'], kind='kde')
```

```python
# Separate histograms for weight and height and age

sns.distplot(a = players_20['height_cm'], label='height', kde=False)
sns.distplot(a = players_20['weight_kg'], label='weight', kde=False)
sns.distplot(a = players_20['age'], label='age', kde=False)

# add title
plt.title("Age, Weight, and Height have completely different distributions")

# Force legend to appear
plt.legend()
```
