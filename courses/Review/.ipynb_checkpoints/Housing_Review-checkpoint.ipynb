```python


# try stratified sampling based on GarageCars, which would affect sale price

# turn GarageCars from integer to discrete variable

# Check GarageCars into discrete strata

train_set['GarageCars'].hist()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=13)

for train_index, val_index in split.split(train, train['GarageCars']):
    train_strat = train.loc[train_index]
    val_strat = train.loc[val_index]
```

```python
train_strat.head()
```

```python
train_strat['GarageCars'].value_counts() / len(train_strat)
```

```python
# compare sampling bias between stratified and purely random sampling

def garage_cat_proportions(data):
    return data['GarageCars'].value_counts() / len(data)

train_set, val_set = train_test_split(train, test_size=0.2, random_state=13)

# compare proportions
compare_proportions = pd.DataFrame({
    "Overall": garage_cat_proportions(train),
    "Stratified": garage_cat_proportions(train_strat),
    "Random": garage_cat_proportions(val_set)
}).sort_index()

# add two error columns
compare_proportions['Rand. %error'] = 100 * compare_proportions['Random'] / compare_proportions['Overall'] - 100
compare_proportions['Strat. %error'] = 100 * compare_proportions['Stratified'] / compare_proportions['Overall'] - 100
```

```python
# You can see Stratified Sampling - based on GarageCars - has lower error than purely Random Sampling
compare_proportions
```

# Data Cleaning

## Numeric Data

```python
# Use a pipeline to deal with both categorical data and columns with missing values

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

# Preprocessing for numerical data (n = 38; int64: 35, float64: 3)
numerical_imputer = SimpleImputer(strategy='median')

# median strategy can only be computed on numerical attributes
# so select all numeric columns from new_train, assign to new_train_numeric
new_train_numeric = new_train.select_dtypes([np.number])

# fit imputer instance to new_train (training) data
# note: fit() ONLY with training data
# fit calculates median and save as internal object

numerical_imputer.fit(new_train_numeric)
```

```python
new_train_numeric.median().values
```

```python
# use 'trained' imputer to transform training set by replacing missing values with learned medians
# transform (after .fit()) will apply transformation to a particular dataset
# X should be a NumPy array containing transformed features

X = numerical_imputer.transform(new_train_numeric)
X[:5]
```

```python
# put X into a DataFrame
# tr = transformed

new_train_tr = pd.DataFrame(X, columns=new_train_numeric.columns,
                           index=new_train_numeric.index)

new_train_tr.head()
```

```python
# interestingly: no missing values among any numeric columns

new_train_tr.isnull().sum()
```

## Categorical Data

```python
new_train_categorical = new_train.select_dtypes(['object'])
new_train_categorical.head(5)
```

```python
new_train_categorical.isnull().sum()
```

```python
# keep it simple, drop all categorical columns with missing values

# get names of columns from new_train_categorical with missing values
cols_with_missing = [col for col in new_train_categorical.columns
                    if new_train_categorical[col].isnull().any()]



cols_with_missing
```

```python
# drop columns in training and validation data
# axis=1 references columns

reduced_new_train_categorical = new_train_categorical.drop(cols_with_missing, axis=1)
reduced_new_val = new_val.drop(cols_with_missing, axis=1)
```

```python
reduced_new_train_categorical.shape
reduced_new_val.shape
```

```python
# Convert from categorical values to one-hot vectors
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
new_train_categorical_1hot = cat_encoder.fit_transform(reduced_new_train_categorical)

# output should be SciPy sparse matrix for more efficient memory usage
new_train_categorical_1hot
```

```python
# got rid of categorical columns with missing values

reduced_new_train_categorical.isnull().sum()
```

# Transformation Pipeline


```python
#review: 

# Training 
# numeric data:        new_train_tr
# categorical data:    reduced_new_train_categorical

# Validation data:     reduced_new_val
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('std_scaler', StandardScaler())
])

new_train_tr_pipe = numeric_pipeline.fit_transform(new_train_tr)
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(new_train_tr)
cat_attribs = list(reduced_new_train_categorical)

full_pipeline = ColumnTransformer([
    ('num', numeric_pipeline, num_attribs),
    ('cat', OneHotEncoder(), cat_attribs)
])

train_prepared = full_pipeline.fit_transform(new_train)
train_prepared
```
